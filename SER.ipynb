```python
##########################################
#          IMPORT THE LIBRARIES          #
##########################################
import os
import sys
import shutil
import timeit
import multiprocessing as mp
import warnings

import numpy as np
import pandas as pd

import librosa
import librosa.display
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.display import Audio, display

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Bidirectional, LSTM, Dense, Dropout, Flatten

from joblib import Parallel, delayed

# Suppress warnings
if not sys.warnoptions:
    warnings.simplefilter("ignore")
warnings.filterwarnings("ignore", category=DeprecationWarning)

print("Libraries imported successfully.")

##########################################
#        DOWNLOAD & SETUP DATASET        #
##########################################
import kagglehub

# Download the RAVDESS Emotional Speech Audio dataset
ravdess = kagglehub.dataset_download("uwrfkaggler/ravdess-emotional-speech-audio")
print("Dataset downloaded from Kagglehub:", ravdess)

# List directories in the downloaded dataset
ravdess_directory_list = os.listdir(ravdess)
print("Files in the dataset:", ravdess_directory_list)
print("Number of directories in dataset:", len(ravdess_directory_list))

# Create a target directory (e.g., in /content) and copy the dataset over
target_dir = '/content/ravdess_dataset'
os.makedirs(target_dir, exist_ok=True)
shutil.copytree(ravdess, target_dir, dirs_exist_ok=True)
ravdess_files = os.listdir(target_dir)
print("Files in the new ravdess_dataset folder:", ravdess_files)

##########################################
#         PROCESS AUDIO FILES            #
##########################################
file_emotion = []
file_path = []

# Iterate through each folder (each actor's folder) in the dataset
for folder in ravdess_directory_list:
    actor_path = os.path.join(ravdess, folder)
    if os.path.isdir(actor_path):
        actor_files = os.listdir(actor_path)
        for f in actor_files:
            full_path = os.path.join(actor_path, f)
            if os.path.isfile(full_path):
                parts = f.split('.')[0].split('-')
                if len(parts) > 2:
                    try:
                        file_emotion.append(int(parts[2]))  # emotion label is the 3rd part
                        file_path.append(full_path)
                    except ValueError:
                        print(f"Skipping file with invalid emotion label: {f}")
                else:
                    print(f"Skipping file due to unexpected structure: {f}")

print(f"Total valid files processed: {len(file_path)}")
print("Sample file paths:", file_path[:5])
print("Sample emotion labels:", file_emotion[:5])

# Debug prints from the last processed file (if available)
if parts and len(parts) > 2:
    print("Last parsed file parts:", parts)
    try:
        print("Parsed emotion as integer:", int(parts[2]))
    except Exception as e:
        print("Error parsing emotion:", e)
    print("Full path of last file:", file_path[-1])
    print("Filename:", f)

##########################################
#         CREATE THE DATAFRAME           #
##########################################
# Build DataFrames for the emotion labels and file paths
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])
path_df = pd.DataFrame(file_path, columns=['Path'])
ravdess_df = pd.concat([emotion_df, path_df], axis=1)

# Map integer labels to actual emotion names
emotion_map = {
    1: 'neutral', 2: 'calm', 3: 'happy', 4: 'sad',
    5: 'angry', 6: 'fear', 7: 'disgust', 8: 'surprise'
}
ravdess_df['Emotions'].replace(emotion_map, inplace=True)

print("\nEmotion value counts from dataset:")
print(ravdess_df['Emotions'].value_counts())
print("First few rows of the DataFrame:")
print(ravdess_df.head())

##########################################
#           PLOT EMOTION COUNTS          #
##########################################
sns.set_style('whitegrid')

plt.figure(figsize=(8, 4))
plt.title('Count of Emotions (Dataset)', size=16)
sns.countplot(x='Emotions', data=ravdess_df,
              order=ravdess_df['Emotions'].value_counts().index, palette='viridis')
plt.ylabel('Count', size=12)
plt.xlabel('Emotions', size=12)
sns.despine(top=True, right=True, left=False, bottom=False)
print("Displaying bar chart: Count of Emotions (Dataset)")
plt.show()

##########################################
#        DEBUG: SAMPLE DATAFRAME         #
##########################################
# (For debugging purposes – a simple example DataFrame)
data_debug = {'Emotions': ['happy', 'sad', 'angry', 'happy', 'neutral',
                           'sad', 'happy', 'angry', 'neutral', 'happy']}
debug_df = pd.DataFrame(data_debug)

plt.figure(figsize=(8, 4))
plt.title('Count of Emotions (Debug Sample)', size=16)
sns.countplot(x='Emotions', data=debug_df,
              order=debug_df['Emotions'].value_counts().index, palette='viridis')
plt.ylabel('Count', size=12)
plt.xlabel('Emotions', size=12)
sns.despine(top=True, right=True, left=False, bottom=False)
print("Displaying bar chart: Count of Emotions (Debug Sample)")
plt.show()

##########################################
#       AUDIO FILE & VISUALIZATION       #
##########################################
# Load the first audio file from the dataset
y, sr = librosa.load(file_path[0])
print(f"Audio sample loaded from: {file_path[0]}")
print(f"Audio sample (snippet): {y[50000:50010]}")
print(f"Audio sample shape: {y.shape}")
print(f"Sample rate: {sr}")

# Play the audio file (if supported in your environment)
print("Playing original audio...")
display(Audio(file_path[0], autoplay=True))

# Plot raw audio waveform
plt.figure(figsize=(10, 5))
plt.title('Raw Audio Example')
pd.Series(y).plot(lw=1)
print("Displaying plot: Raw Audio Example")
plt.show()

# Trim silence from the audio and plot the trimmed waveform
y_trimmed, _ = librosa.effects.trim(y, top_db=20)
plt.figure(figsize=(10, 5))
plt.title('Raw Audio Trimmed Example')
pd.Series(y_trimmed).plot(lw=1, color='green')
print("Displaying plot: Raw Audio Trimmed Example")
plt.show()

# Zoom in on a segment of the raw audio and plot
plt.figure(figsize=(10, 5))
plt.title('Raw Audio Zoomed In Example')
pd.Series(y[30000:31000]).plot(lw=1, color='red')
print("Displaying plot: Raw Audio Zoomed In Example")
plt.show()

# Compute and display the spectrogram (STFT)
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
print("Spectrogram shape:", S_db.shape)
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db, x_axis='time', y_axis='log', ax=ax)
ax.set_title('Spectrogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format='%0.0f')
print("Displaying plot: Spectrogram Example")
plt.show()

# Compute and display the Mel spectrogram
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=256)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db_mel, x_axis='time', y_axis='log', ax=ax)
ax.set_title('Mel Spectrogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format='%0.2f')
print("Displaying plot: Mel Spectrogram Example")
plt.show()

##########################################
#        SAVE DATAFRAME TO CSV           #
##########################################
data_path = ravdess_df.copy()  # You can use the same DataFrame
data_path.to_csv("data_path.csv", index=False)
print("Saved data_path.csv")
print("Counts of emotions in data_path:")
print(data_path['Emotions'].value_counts())

# Plot count of emotions using the saved DataFrame
plt.figure(figsize=(8, 4))
plt.title('Count of Emotions (data_path)', size=16)
sns.countplot(x='Emotions', data=data_path,
              order=data_path['Emotions'].value_counts().index, palette='viridis')
plt.ylabel('Count', size=12)
plt.xlabel('Emotions', size=12)
sns.despine(top=True, right=True, left=False, bottom=False)
print("Displaying bar chart: Count of Emotions (data_path)")
plt.show()

##########################################
#       AUDIO AUGMENTATION EXAMPLES      #
##########################################
# Define audio augmentation functions

def pitch(data, sr, n_steps=2):
    return librosa.effects.pitch_shift(y=data, sr=sr, n_steps=n_steps)

def noise(data):
    noise_amp = 0.035 * np.random.uniform() * np.amax(data)
    return data + noise_amp * np.random.normal(size=data.shape[0])

def stretch(data, rate=0.8):
    return librosa.effects.time_stretch(data, rate=rate)

def shift(data):
    shift_range = int(np.random.uniform(low=-5, high=5) * 1000)
    return np.roll(data, shift_range)

# Plot and play original waveform
plt.figure(figsize=(12, 5))
plt.title('Original Audio Waveform')
librosa.display.waveshow(y=y, sr=sr)
print("Displaying plot: Original Audio Waveform")
plt.show()
print("Playing original audio...")
display(Audio(y, rate=sr))

# Apply noise and display
y_noise = noise(y)
plt.figure(figsize=(12, 5))
plt.title('Audio with Noise')
librosa.display.waveshow(y=y_noise, sr=sr)
print("Displaying plot: Audio with Noise")
plt.show()
print("Playing audio with noise...")
display(Audio(y_noise, rate=sr))

# Apply stretch and display
y_stretch = stretch(y, rate=0.5)
plt.figure(figsize=(12, 5))
plt.title('Stretched Audio')
librosa.display.waveshow(y=y_stretch, sr=sr)
print("Displaying plot: Stretched Audio")
plt.show()
print("Playing stretched audio...")
display(Audio(y_stretch, rate=sr))

# Apply shift and display
y_shift = shift(y)
plt.figure(figsize=(12, 5))
plt.title('Shifted Audio')
librosa.display.waveshow(y=y_shift, sr=sr)
print("Displaying plot: Shifted Audio")
plt.show()
print("Playing shifted audio...")
display(Audio(y_shift, rate=sr))

# Apply pitch shift and display
y_pitch = pitch(y, sr, n_steps=-5)
plt.figure(figsize=(12, 5))
plt.title('Audio with Pitch Shift')
librosa.display.waveshow(y=y_pitch, sr=sr)
print("Displaying plot: Audio with Pitch Shift")
plt.show()
print("Playing audio with pitch shift...")
display(Audio(y_pitch, rate=sr))

##########################################
#       FEATURE EXTRACTION FUNCTIONS     #
##########################################
def extract_features(data, sr, frame_length=2048, hop_length=512):
    # Compute Short-Time Fourier Transform
    stft_data = np.abs(librosa.stft(data, n_fft=frame_length, hop_length=hop_length))
    # Zero-Crossing Rate
    zcr = librosa.feature.zero_crossing_rate(data, frame_length=frame_length, hop_length=hop_length).squeeze()
    # Root Mean Square Energy (RMS)
    rmse = librosa.feature.rms(S=stft_data).squeeze()
    # Mel Spectrogram
    mel_spec = librosa.feature.melspectrogram(y=data, sr=sr, n_fft=frame_length, hop_length=hop_length)
    # Mel-Frequency Cepstral Coefficients (MFCC)
    mfcc = librosa.feature.mfcc(S=librosa.power_to_db(mel_spec), sr=sr, n_mfcc=13).T.ravel()
    # Combine all features into one array
    return np.hstack([zcr, rmse, mfcc])

def augment_and_extract(data, sr, frame_length=2048, hop_length=512):
    augments = [
        data,
        data + 0.005 * np.random.normal(0, 1, len(data)),                    # Original + noise
        librosa.effects.pitch_shift(data, sr=sr, n_steps=2),                   # Pitched
        librosa.effects.pitch_shift(data + 0.005 * np.random.normal(0, 1, len(data)),
                                    sr=sr, n_steps=2)                        # Pitched + noise
    ]
    return np.vstack([extract_features(aug, sr, frame_length, hop_length) for aug in augments])

def get_features(path, duration=2.5, offset=0.6, sr=22050, frame_length=2048, hop_length=512):
    data, sr = librosa.load(path, duration=duration, offset=offset, sr=sr)
    return augment_and_extract(data, sr, frame_length, hop_length)

print("Number of processors available:", mp.cpu_count())

##########################################
#      PARALLEL FEATURE EXTRACTION       #
##########################################
start = timeit.default_timer()

def process_feature(path, emotion):
    features = get_features(path)
    X = []
    Y = []
    # Four augmentations per file
    for ele in features:
        X.append(ele)
        Y.append(emotion)
    return X, Y

paths = data_path['Path']
emotions = data_path['Emotions']

results = Parallel(n_jobs=-1)(delayed(process_feature)(path, emotion)
                              for (path, emotion) in zip(paths, emotions))
X_features = []
Y_labels = []
for x_vals, y_vals in results:
    X_features.extend(x_vals)
    Y_labels.extend(y_vals)
stop = timeit.default_timer()
print('Feature extraction time:', stop - start)
print("Number of extracted feature samples:", len(X_features))
print("Number of labels:", len(Y_labels))

# Save the extracted features in a CSV file
features_df = pd.DataFrame(X_features)
features_df['Emotions'] = Y_labels
features_df.to_csv('emotion.csv', index=False)
print("Saved extracted features to emotion.csv")
print("First few rows of extracted features:")
print(features_df.head())

# Check and fill missing values
print("Missing values in features:\n", features_df.isna().sum())
features_df.fillna(0, inplace=True)
print("Total missing values after fill:", features_df.isna().sum().sum())

##########################################
#       PREPARE DATA FOR TRAINING        #
##########################################
# Separate features and labels
X = features_df.iloc[:, :-1].values
Y = features_df['Emotions'].values

# One-hot encode labels for multiclass classification
encoder = OneHotEncoder()
Y_encoded = encoder.fit_transform(np.array(Y).reshape(-1, 1)).toarray()

x_train, x_test, y_train, y_test = train_test_split(X, Y_encoded, random_state=42,
                                                    test_size=0.2, shuffle=True)
print("Training data shapes:", x_train.shape, y_train.shape)
print("Testing data shapes:", x_test.shape, y_test.shape)

# Reshape for Conv1D input (add channel dimension)
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test  = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)

##########################################
#         BUILD & TRAIN THE MODEL        #
##########################################
model = tf.keras.Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu',
           input_shape=(x_train.shape[1], x_train.shape[2])),
    MaxPooling1D(pool_size=2),
    Dropout(0.2),
    Conv1D(filters=128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Dropout(0.3),
    Bidirectional(LSTM(128, return_sequences=True)),
    Bidirectional(LSTM(64)),
    Dense(128, activation='relu'),
    Dropout(0.4),
    Dense(8, activation='softmax')  # 8 emotion classes
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print("Model summary:")
model.summary()

history = model.fit(x_train, y_train, validation_data=(x_test, y_test),
                    epochs=30, batch_size=128)

##########################################
#         PLOT TRAINING METRICS          #
##########################################
plt.figure(figsize=(8, 5))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy over Epochs')
print("Displaying plot: Model Accuracy over Epochs")
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss over Epochs')
print("Displaying plot: Model Loss over Epochs")
plt.show()

##########################################
#            SAVE THE MODEL              #
##########################################
model.save('emotion_recognition_model.h5')
print("Model saved as 'emotion_recognition_model.h5'")
```

```python
from google.colab import files
import tensorflow as tf
import numpy as np
import librosa

# Load your pre-trained model
model = tf.keras.models.load_model('emotion_recognition_model.h5')

# Define emotion labels (update these if your training labels differ)
EMOTIONS = ['angry', 'happy', 'sad', 'neutral', 'fear', 'surprise', 'disgust']

def extract_features(file_path):
    """
    Loads an audio file, extracts MFCC features, pads/truncates them to a fixed size,
    flattens the matrix, and reshapes the result to match the training input.
    """
    # Load audio file with librosa
    audio, sample_rate = librosa.load(file_path, sr=22050, duration=3)

    # Extract MFCC features (n_mfcc should match training settings)
    mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)

    # Pad or truncate along the time axis (we want a fixed number of frames, e.g., 300)
    max_pad_len = 300
    if mfccs.shape[1] < max_pad_len:
        pad_width = max_pad_len - mfccs.shape[1]
        mfccs = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)), mode='constant')
    else:
        mfccs = mfccs[:, :max_pad_len]

    # Flatten the MFCCs to create a 1D feature vector.
    # For example, a (40, 300) matrix becomes a vector of length 40*300 = 12000.
    features = mfccs.flatten()

    # Reshape to (1, feature_length, 1) to match the Conv1D model input.
    features = features.reshape(1, -1, 1)
    return features

# Upload an audio file using the file upload widget
uploaded = files.upload()

# Process each uploaded file and make predictions
for file_name in uploaded.keys():
    try:
        # Extract features from the audio file
        features = extract_features(file_name)

        # Predict using the loaded model
        predictions = model.predict(features)
        predicted_emotion = EMOTIONS[np.argmax(predictions)]

        # Print out the results
        print(f"\nFile: {file_name}")
        print(f"Predicted emotion: {predicted_emotion}")
        print("Confidence scores:")
        for emotion, score in zip(EMOTIONS, predictions[0]):
            print(f"  {emotion}: {score:.4f}")

    except Exception as e:
        print(f"Error processing {file_name}: {str(e)}")
```
